# Import the necesaary library and load the dataset

```python
import pandas as pd

# Load generation data
generation_df = pd.read_csv("generation_data.csv")

# Load demand dataset
demand_df = pd.read_csv("demand_data.csv")

demand_df.head(5)
```

# Convert datatype to a format suitable for analysis and check types - Demand Dataset

```python
# Convert 'Date' to datetime with the correct format
demand_df['Date'] = pd.to_datetime(demand_df['Date'], format='%d/%m/%Y')

# Convert 'Hourly Profile' to text (string)
demand_df['Hourly Profile'] = demand_df['Hourly Profile'].astype(str)

# Convert 'Season' to text (string)
demand_df['Season'] = demand_df['Season'].astype(str)

# Convert 'Date (MST)' to datetime
demand_df['Date (MST)'] = pd.to_datetime(demand_df['Date (MST)'], format='%d/%m/%Y %H:%M:%S')
# Convert 'Region' to text (string)
demand_df['Region'] = demand_df['Region'].astype(str)

# Convert 'Load' to float
demand_df['Load'] = demand_df['Load'].astype(float)

# Verify the changes
print(demand_df.dtypes)
```

# Convert datatype to a format suitable for analysis and check data types - Generation Dataset

```python
# Convert 'Fuel Type' to string
generation_df['Fuel Type'] = generation_df['Fuel Type'].astype(str)

# Convert 'Date' to datetime with the correct format
generation_df['Date'] = pd.to_datetime(generation_df['Date'], format='%d/%m/%Y')

# Convert 'Date (MST)' to datetime with the correct format
generation_df['Date (MST)'] = pd.to_datetime(generation_df['Date (MST)'], format='%d/%m/%Y %H:%M:%S')

# Create a new DataFrame with the modified columns
generation_df = pd.DataFrame({
    'Fuel Type': generation_df['Fuel Type'],
    'Date': generation_df['Date'],
    'Date (MST)': generation_df['Date (MST)'],
    'Maximum Capacity': generation_df['Maximum Capacity'].astype(float),
    'System Available': generation_df['System Available'].astype(float),
    'System Capacity': generation_df['System Capacity'].astype(float),
    'System Generation': generation_df['System Generation'].astype(float)
})

# Display the updated Datatypes
print(generation_df.dtypes)
```

# Daily System Generation vs. Daily Total Load for last 2 years

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter data for the last 2 years
start_date = pd.to_datetime('today') - pd.DateOffset(years=2)
df_last_3_years = demand_df[demand_df['Date'] >= start_date]
generation_df_last_3_years = generation_df[generation_df['Date'] >= start_date]

# Calculate daily total load and daily total system generation
daily_total_load = df_last_3_years.groupby(df_last_3_years['Date'].dt.date)['Load'].sum().reset_index()
daily_total_generation = generation_df_last_3_years.groupby(generation_df_last_3_years['Date'].dt.date)['System Generation'].sum().reset_index()

# Merge the two dataframes on the 'Date' column
merged_df = pd.merge(daily_total_load, daily_total_generation, on='Date', how='outer')

# Plot daily system generation vs. total load with line graph
plt.figure(figsize=(12, 6))

plt.plot(merged_df['Date'], merged_df['Load'], label='Daily Total Load')
plt.plot(merged_df['Date'], merged_df['System Generation'], label='Daily Total Energy Generation')

plt.title('Daily Energy Generation vs. Daily Total Load for the last 2 years')
plt.xlabel('Date')
plt.ylabel('Daily Energy (MWh)')
plt.legend()
plt.show()
```

# Monthly Demand Trend for Each Regions Over last 5 Years

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert 'Date' column to datetime format
demand_df['Date'] = pd.to_datetime(demand_df['Date'])

# Filter data for the last 5 years
end_date = demand_df['Date'].max()
start_date = end_date - pd.DateOffset(years=5)
demand_df_last_5_years = demand_df[(demand_df['Date'] >= start_date) & (demand_df['Date'] <= end_date)].copy()

# Extract year and month from the date column using .loc to avoid SettingWithCopyWarning
demand_df_last_5_years.loc[:, 'Year'] = demand_df_last_5_years['Date'].dt.year
demand_df_last_5_years.loc[:, 'Month'] = demand_df_last_5_years['Date'].dt.month

# Group data by region and month to calculate total demand
demand_by_region_month = demand_df_last_5_years.groupby(['Region', 'Year', 'Month'])['Load'].sum().unstack(level=0)

# Convert the index (year, month tuples) to string representation
demand_by_region_month.index = [f'{year}-{month:02d}' for year, month in demand_by_region_month.index]

# Plot demand trend for each region over the months
plt.figure(figsize=(12, 6))
for region in demand_by_region_month.columns:
    plt.plot(demand_by_region_month.index, demand_by_region_month[region], label=region)

plt.title('Demand Trend for Each Region Over the Months (Last 5 Years)', fontsize=16)
plt.xlabel('Year-Month', fontsize=14)
plt.ylabel('Total Demand', fontsize=14)
plt.xticks(rotation=45, fontsize=10)  # Rotate x-axis labels and reduce font size
plt.yticks(fontsize=10)  # Reduce font size for y-axis labels
plt.legend(fontsize=10)

# Set y-axis limits and increase padding
plt.ylim(bottom=0)  # Set the lower limit to 0
plt.tight_layout()  # Adjust padding for labels

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load demand data
demand_df = pd.read_csv("demand_data.csv")

# Convert 'Date' column to datetime format with the correct format for demand data
demand_df['Date'] = pd.to_datetime(demand_df['Date'], format='%d/%m/%Y')

# Set 'Date' column as index for demand data
demand_df.set_index('Date', inplace=True)

# Resample the data to daily frequency for demand data
demand_daily = demand_df.resample('D').sum()

# Split the data into training and testing sets
train_size = int(len(demand_daily) * 0.8)
train_demand, test_demand = demand_daily.iloc[:train_size], demand_daily.iloc[train_size:]

# Fit ARIMA model for demand data
model_demand = ARIMA(train_demand['Load'], order=(5,1,0)) # Example order, you may need to tune this
model_fit_demand = model_demand.fit()

# Forecast for demand data
forecast_demand = model_fit_demand.forecast(steps=len(test_demand)) # Forecast for the test set

# Plot the training, testing, and prediction
plt.figure(figsize=(12, 6))
plt.plot(train_demand.index, train_demand['Load'], label='Training Data')
plt.plot(test_demand.index, test_demand['Load'], label='Testing Data')
plt.plot(test_demand.index, forecast_demand, label='Prediction')
plt.title('Demand Forecasting with ARIMA')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load demand data
demand_df = pd.read_csv("demand_data.csv")

# Convert 'Date' column to datetime format with the correct format for demand data
demand_df['Date'] = pd.to_datetime(demand_df['Date'], format='%d/%m/%Y')

# Set 'Date' column as index for demand data
demand_df.set_index('Date', inplace=True)

# Resample the data to daily frequency for demand data
demand_daily = demand_df.resample('D').sum()

# Split the data into training and testing sets
train_size = int(len(demand_daily) * 0.8)
train_demand, test_demand = demand_daily.iloc[:train_size], demand_daily.iloc[train_size:]

# Fit ARIMA model for demand data
model_demand = ARIMA(train_demand['Load'], order=(5,1,0)) # Example order, you may need to tune this
model_fit_demand = model_demand.fit()

# Forecast for demand data for the years 2024 and 2025
forecast_demand = model_fit_demand.forecast(steps=365 * 2) # Forecast for 2 years (2024 and 2025)

# Plot the training, testing, and prediction
plt.figure(figsize=(12, 6))
plt.plot(train_demand.index, train_demand['Load'], label='Training Data')
plt.plot(test_demand.index, test_demand['Load'], label='Testing Data')
plt.plot(pd.date_range(start=test_demand.index[0], periods=len(forecast_demand)), forecast_demand, label='Prediction')
plt.title('Demand Forecasting with ARIMA')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load demand and generation data
demand_df = pd.read_csv("demand_data.csv")
generation_df = pd.read_csv("generation_data.csv")

# Convert 'Date' column to datetime format with the correct format for both demand and generation data
demand_df['Date'] = pd.to_datetime(demand_df['Date'], format='%d/%m/%Y')
generation_df['Date'] = pd.to_datetime(generation_df['Date'], format='%d/%m/%Y')

# Set 'Date' column as index for both demand and generation data
demand_df.set_index('Date', inplace=True)
generation_df.set_index('Date', inplace=True)

# Resample the data to daily frequency for both demand and generation data
demand_daily = demand_df.resample('D').sum()
generation_daily = generation_df.resample('D').sum()

# Split the data into training and testing sets for demand data
train_size = int(len(demand_daily) * 0.8)
train_demand, test_demand = demand_daily.iloc[:train_size], demand_daily.iloc[train_size:]

# Split the data into training and testing sets for generation data
train_generation, test_generation = generation_daily.iloc[:train_size], generation_daily.iloc[train_size:]

# Fit ARIMA model for demand data
model_demand = ARIMA(train_demand['Load'], order=(5,1,0)) # Example order, you may need to tune this
model_fit_demand = model_demand.fit()

# Fit ARIMA model for generation data
model_generation = ARIMA(train_generation['System Generation'], order=(5,1,0)) # Example order, you may need to tune this
model_fit_generation = model_generation.fit()

# Forecast for demand data
forecast_demand = model_fit_demand.forecast(steps=len(test_demand))

# Forecast for generation data
forecast_generation = model_fit_generation.forecast(steps=len(test_generation))

# Plot the training, testing, and prediction for both demand and generation data
plt.figure(figsize=(12, 6))
plt.plot(train_demand.index, train_demand['Load'], label='Training Demand Data', color='blue')
plt.plot(test_demand.index, test_demand['Load'], label='Testing Demand Data', color='green')
plt.plot(test_demand.index, forecast_demand, label='Demand Prediction', color='red')

plt.plot(train_generation.index, train_generation['System Generation'], label='Training Generation Data', color='cyan')
plt.plot(test_generation.index, test_generation['System Generation'], label='Testing Generation Data', color='magenta')
plt.plot(test_generation.index, forecast_generation, label='Generation Prediction', color='yellow')

plt.title('Demand and Generation Forecasting with ARIMA')
plt.xlabel('Date')
plt.ylabel('Demand/Generation')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA

# Load demand and generation data
demand_df = pd.read_csv("demand_data.csv")
generation_df = pd.read_csv("generation_data.csv")

# Convert 'Date' column to datetime format with the correct format for both demand and generation data
demand_df['Date'] = pd.to_datetime(demand_df['Date'], format='%d/%m/%Y')
generation_df['Date'] = pd.to_datetime(generation_df['Date'], format='%d/%m/%Y')

# Set 'Date' column as index for both demand and generation data
demand_df.set_index('Date', inplace=True)
generation_df.set_index('Date', inplace=True)

# Extract data from 2021 onwards
demand_df = demand_df['2021':]
generation_df = generation_df['2021':]

# Resample the data to daily frequency for both demand and generation data
demand_daily = demand_df.resample('D').sum()
generation_daily = generation_df.resample('D').sum()

# Train ARIMA models for demand and generation data using data up to 2023
model_demand = ARIMA(demand_daily['Load'].loc[:'2023-12-31'], order=(5, 1, 0))
model_generation = ARIMA(generation_daily['System Generation'].loc[:'2023-12-31'], order=(5, 1, 0))

# Fit ARIMA models
model_fit_demand = model_demand.fit()
model_fit_generation = model_generation.fit()

# Forecast demand and generation for 2024 to 2026
forecast_demand = model_fit_demand.forecast(steps=365 * 3)
forecast_generation = model_fit_generation.forecast(steps=365 * 3)

# Plot the original data and forecast with seasonality for demand
plt.figure(figsize=(12, 6))
plt.plot(demand_daily.index, demand_daily['Load'], label='Original Demand Data')
plt.plot(pd.date_range(start=demand_daily.index[-1], periods=365*3), forecast_demand, label='Demand Forecast', color='red')

# Add seasonal component to the demand forecast
decomposition_demand = seasonal_decompose(forecast_demand, model='additive')
seasonal_demand = decomposition_demand.seasonal
plt.plot(pd.date_range(start=demand_daily.index[-1], periods=365*3), seasonal_demand, label='Demand Seasonality', linestyle='--', color='green')

plt.title('Demand Forecasting with ARIMA including Seasonality')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.legend()
plt.show()

# Plot the original data and forecast with seasonality for generation
plt.figure(figsize=(12, 6))
plt.plot(generation_daily.index, generation_daily['System Generation'], label='Original Generation Data')
plt.plot(pd.date_range(start=generation_daily.index[-1], periods=365*3), forecast_generation, label='Generation Forecast', color='blue')

# Add seasonal component to the generation forecast
decomposition_generation = seasonal_decompose(forecast_generation, model='additive')
seasonal_generation = decomposition_generation.seasonal
plt.plot(pd.date_range(start=generation_daily.index[-1], periods=365*3), seasonal_generation, label='Generation Seasonality', linestyle='--', color='orange')

plt.title('Generation Forecasting with ARIMA including Seasonality')
plt.xlabel('Date')
plt.ylabel('Generation')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA

# Load demand and generation data
demand_df = pd.read_csv("demand_data.csv")
generation_df = pd.read_csv("generation_data.csv")

# Convert 'Date' column to datetime format with the correct format for both demand and generation data
demand_df['Date'] = pd.to_datetime(demand_df['Date'], format='%d/%m/%Y')
generation_df['Date'] = pd.to_datetime(generation_df['Date'], format='%d/%m/%Y')

# Set 'Date' column as index for both demand and generation data
demand_df.set_index('Date', inplace=True)
generation_df.set_index('Date', inplace=True)

# Extract data from 2021 onwards
demand_df = demand_df['2021':]
generation_df = generation_df['2021':]

# Resample the data to daily frequency for both demand and generation data
demand_daily = demand_df.resample('D').sum()
generation_daily = generation_df.resample('D').sum()

# Train ARIMA models for demand and generation data using data up to 2023
model_demand = ARIMA(demand_daily['Load'].loc[:'2023-12-31'], order=(5, 1, 0))
model_generation = ARIMA(generation_daily['System Generation'].loc[:'2023-12-31'], order=(5, 1, 0))

# Fit ARIMA models
model_fit_demand = model_demand.fit()
model_fit_generation = model_generation.fit()

# Forecast demand and generation for 2024 to 2026
forecast_demand = model_fit_demand.forecast(steps=365 * 3)
forecast_generation = model_fit_generation.forecast(steps=365 * 3)

# Perform seasonal decomposition for the demand forecast
decomposition_demand = seasonal_decompose(forecast_demand, model='additive')
seasonal_demand = decomposition_demand.seasonal

# Perform seasonal decomposition for the generation forecast
decomposition_generation = seasonal_decompose(forecast_generation, model='additive')
seasonal_generation = decomposition_generation.seasonal

# Plot the original data and forecast with seasonality for both demand and generation
plt.figure(figsize=(12, 6))

# Plot demand data
plt.plot(demand_daily.index, demand_daily['Load'], label='Original Demand Data', color='blue')
plt.plot(pd.date_range(start=demand_daily.index[-1], periods=365*3), forecast_demand + seasonal_demand, label='Demand Forecast', color='red')

# Plot generation data
plt.plot(generation_daily.index, generation_daily['System Generation'], label='Original Generation Data', color='orange')
plt.plot(pd.date_range(start=generation_daily.index[-1], periods=365*3), forecast_generation + seasonal_generation, label='Generation Forecast', color='purple')

plt.title('Demand and Generation Forecasting with ARIMA including Seasonality')
plt.xlabel('Date')
plt.ylabel('Demand/Generation')
plt.legend()
plt.show()
```
